# Employee Attrition

## Introduction

In business, employee attrition is when employees leave the company for whatever reason, either they've found a new job or retired, and haven't been replaced immediately.


For a company to be successful, it needs not only to attract top talent but it also needs to retain these talents. For this reason, I have the job to look into a dataset containing information regarding a certain company's employee list to try to find patterns that may provide useful information in understanding why employees leave.

After an EDA, I'll build a machine learning model for employee attrition prediction.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import joblib

# 1. Reload Data (to be safe)
df = pd.read_csv("data.csv")

# 2. Select Key Features (Added Distance, WorkLifeBalance, EnvironmentSatisfaction)
selected_features = [
    'Age', 
    'MonthlyIncome', 
    'OverTime', 
    'JobSatisfaction', 
    'YearsAtCompany', 
    'TotalWorkingYears',
    'DistanceFromHome',      # NEW
    'WorkLifeBalance',       # NEW
    'EnvironmentSatisfaction' # NEW
]

# 3. Manual Encoding (Safe & Consistent for Apps)
# We map "Yes/No" to 1/0 manually so the App knows exactly what to send
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})

X_demo = df[selected_features].copy()
y_demo = df['Attrition'].map({'Yes': 1, 'No': 0})

# 4. Train the Enhanced Model
model_demo = RandomForestClassifier(n_estimators=100, random_state=42)
model_demo.fit(X_demo, y_demo)

# 5. Save
joblib.dump(model_demo, 'model_demo.pkl')
print("Enhanced model saved successfully!")
```
